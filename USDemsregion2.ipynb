```python
import pandas as pd
import numpy as np
import json 
```

```python
df= pd.read_csv('USDemsregion.csv')
df
```

```python
del df['Unnamed: 105']
del df['Unnamed: 106']
del df['Unnamed: 107']
del df['ID']
df
```

Unstacking so all columns merged into one column 'region'.

```python

df = df.unstack().to_frame('Region').reset_index(drop=True) 
df
```

```python
df2= df.dropna()
df2
```

Removing all rows which contain information on the percentage/distribution, left with only the names of regions.

```python
df3 = df2[df2['Region'].str.contains('percentage')== False]
df3
```

Splitting the string and using replace to remove surplus symbols/punctuation etc. Produces the name of the State. 

```python
df3['Region'] = df3['Region'].str.split(':').str[1].str.replace('{"region"','').str.replace('"','')
df3
```

Doing a value count to produce a value which represents the number of ads which have been posted in each State, then converting this list back into a dataframe and renaming columns. 

```python
df4 = df3['Region'].value_counts()
df5 = pd.DataFrame(df4)
df6= df5.reset_index()
df7= df6.rename(columns={'index': 'State', 'Region': 'Number of Ads'})
df7
```

Cleaning data by filtering out the Regions with less than 800 ads to remove anomalies and unknowns.


```python
df8 = df7[df7['Number of Ads']>800]
df8
```

Writing to json. 

```python
USDemsReg = list(df8.T.to_dict().values())
open('USDemsReg.json', 'w').write(json.dumps(USDemsReg))
```

```python
df8.to_csv('USDemsReg.csv')
```
