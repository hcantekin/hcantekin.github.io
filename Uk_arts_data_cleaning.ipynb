```python
import pandas as pd
import numpy as np
```

```python
df= pd.read_csv('creative_industry_data.csv')
df.head()
```

```python
# Convert wide format to long format
df2 = pd.melt(df, id_vars=['region', 'industry'], var_name='date', value_name='value')

df2.head()
```

```python
import json

UK_arts_data = list(df2.T.to_dict().values())

open('UK_arts_data.json','w').write(json.dumps(UK_arts_data))
```
