```python
import pandas as pd
import numpy as np
import json 
from google.colab import files
import os
```

```python
df = pd.read_csv('USDems19_22.csv')
df
```

```python
del df['durn']
```

```python
df1= df.dropna()
df1
```

```python
df1['distribution'] = df1['distribution'].astype('string')
df1['region'] = df1['region'].astype('string')

```

```python
print(df1.dtypes)

df1
```

del df1['Distribution ']

```python

df2= df1[df1['distribution'].str.contains("percentage:1}")]
df2
```

```python

df2['region'] = df2['region'].str.split(':').str[1].str.replace('{"region"','').str.replace('"','')
df2
```

```python
df3 = df2['region'].value_counts()
df4 = pd.DataFrame(df3)
df5= df4.reset_index()
df6= df5.rename(columns={'index': 'State', 'delivery_by_region': 'Number of Ads'})
df6
```

```python
USDems20 = list(df5.T.to_dict().values())
open('USDems20.json', 'w').write(json.dumps(USDems20))
```
